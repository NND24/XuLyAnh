```python
import cv2
import matplotlib.pyplot as plt

# Đọc ảnh gốc
image = cv2.imread("./landscape Images/color/0.jpg")

# Kiểm tra xem ảnh có đọc được không
if image is None:
    print("Error: Không thể đọc được ảnh.")
else:
    # Chuyển đổi ảnh thành grayscale
    grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Đảo ngược ảnh grayscale
    invert = cv2.bitwise_not(grey_image)

    # Làm mờ ảnh đã đảo ngược
    blur = cv2.GaussianBlur(invert, (21, 21), 0)

    # Đảo ngược ảnh đã làm mờ
    inverted_blur = cv2.bitwise_not(blur)

    # Tạo hiệu ứng tranh vẽ bút chì bằng cách chia ảnh grayscale cho ảnh làm mờ đảo ngược
    sketch = cv2.divide(grey_image, inverted_blur, scale=256.0)

    # Hiển thị kết quả dưới dạng hình ảnh
    plt.imshow(sketch, cmap='gray')  # Sử dụng cmap='gray' để hiển thị ảnh dưới dạng grayscale
    plt.axis('off')  # Ẩn trục tọa độ
    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Đọc ảnh gốc
image = cv2.imread("./landscape Images/color/0.jpg")

# Kiểm tra xem ảnh có đọc được không
if image is None:
    print("Error: Không thể đọc được ảnh.")
else:
    # Điều chỉnh kích thước ảnh nhỏ lại (tạo hiệu ứng pixel)
    height, width = image.shape[:2]

    # Giảm độ phân giải (càng nhỏ, ảnh sẽ càng pixelated)
    pixel_size = 16  # Điều chỉnh kích thước pixel theo ý muốn
    temp_image = cv2.resize(image, (width // pixel_size, height // pixel_size), interpolation=cv2.INTER_LINEAR)

    # Phóng to lại ảnh về kích thước ban đầu
    pixelated_image = cv2.resize(temp_image, (width, height), interpolation=cv2.INTER_NEAREST)

    # Chuyển đổi BGR sang RGB để hiển thị với matplotlib
    pixelated_image_rgb = cv2.cvtColor(pixelated_image, cv2.COLOR_BGR2RGB)

    # Hiển thị ảnh
    plt.imshow(pixelated_image_rgb)
    plt.axis('off')  # Ẩn trục tọa độ
    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Đọc ảnh gốc
image = cv2.imread("./landscape Images/color/0.jpg")

if image is None:
    print("Error: Không thể đọc được ảnh.")
else:
    # Chuyển thành grayscale
    grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Làm mờ ảnh để giảm chi tiết
    grey_image = cv2.medianBlur(grey_image, 5)

    # Phát hiện cạnh bằng thuật toán adaptive thresholding
    edges = cv2.adaptiveThreshold(grey_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)

    # Làm mờ và giảm độ sắc nét để tạo hiệu ứng hoạt hình
    color = cv2.bilateralFilter(image, 9, 300, 300)

    # Kết hợp ảnh màu và các cạnh
    cartoon = cv2.bitwise_and(color, color, mask=edges)

    # Chuyển sang RGB để hiển thị
    cartoon_rgb = cv2.cvtColor(cartoon, cv2.COLOR_BGR2RGB)

    # Hiển thị
    plt.imshow(cartoon_rgb)
    plt.axis('off')
    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Đọc ảnh gốc
image = cv2.imread("./landscape Images/color/0.jpg")

if image is None:
    print("Error: Không thể đọc được ảnh.")
else:
    # Tạo ma trận sepia
    sepia_filter = np.array([[0.393, 0.769, 0.189],
                             [0.349, 0.686, 0.168],
                             [0.272, 0.534, 0.131]])

    # Áp dụng hiệu ứng sepia bằng cách nhân ảnh gốc với ma trận sepia
    sepia_image = cv2.transform(image, sepia_filter)
    
    # Giới hạn giá trị pixel từ 0 đến 255
    sepia_image = np.clip(sepia_image, 0, 255).astype(np.uint8)

    # Chuyển sang RGB để hiển thị
    sepia_image_rgb = cv2.cvtColor(sepia_image, cv2.COLOR_BGR2RGB)

    # Hiển thị
    plt.imshow(sepia_image_rgb)
    plt.axis('off')
    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Đọc ảnh gốc
image = cv2.imread("./landscape Images/color/0.jpg")

if image is None:
    print("Error: Không thể đọc được ảnh.")
else:
    # Áp dụng hiệu ứng tăng cường ánh sáng CLAHE
    lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)

    # Tăng cường kênh L bằng CLAHE (Contrast Limited Adaptive Histogram Equalization)
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
    l = clahe.apply(l)

    # Gộp lại thành ảnh
    lab = cv2.merge((l, a, b))
    hdr_image = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)

    # Chuyển BGR sang RGB để hiển thị
    hdr_image_rgb = cv2.cvtColor(hdr_image, cv2.COLOR_BGR2RGB)

    # Hiển thị
    plt.imshow(hdr_image_rgb)
    plt.axis('off')
    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Đọc ảnh gốc
image = cv2.imread("./landscape Images/color/0.jpg")

if image is None:
    print("Error: Không thể đọc được ảnh.")
else:
    # Chuyển ảnh sang grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Làm mịn ảnh bằng Gaussian Blur
    gray_blur = cv2.medianBlur(gray, 5)
    
    # Phát hiện cạnh bằng thuật toán adaptive threshold
    edges = cv2.adaptiveThreshold(
        gray_blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9
    )
    
    # Áp dụng bộ lọc bilateral để làm mịn các vùng màu
    color = cv2.bilateralFilter(image, 9, 300, 300)
    
    # Kết hợp hình ảnh màu và cạnh để tạo hiệu ứng hoạt hình
    cartoon = cv2.bitwise_and(color, color, mask=edges)

    # Chuyển BGR sang RGB để hiển thị
    cartoon_rgb = cv2.cvtColor(cartoon, cv2.COLOR_BGR2RGB)

    # Hiển thị
    plt.imshow(cartoon_rgb)
    plt.axis('off')
    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Đọc ảnh nền và ảnh cần ghép
src = cv2.imread("./landscape Images/color/0.jpg")  # Ảnh cần ghép
dst = cv2.imread("style_3.png")  # Ảnh nền

# Tạo mask từ ảnh nguồn
src_mask = 255 * np.ones(src.shape, src.dtype)

# Tạo điểm giữa ảnh nền, nơi ảnh nguồn sẽ được ghép vào
center = (dst.shape[1] // 2, dst.shape[0] // 2)

# Áp dụng hiệu ứng ghép Seamless Cloning
output = cv2.seamlessClone(src, dst, src_mask, center, cv2.NORMAL_CLONE)

# Chuyển BGR sang RGB để hiển thị
output_rgb = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)

# Hiển thị
plt.imshow(output_rgb)
plt.axis('off')
plt.show()
```
