```python
import cv2
import matplotlib.pyplot as plt

# c nh gc
image = cv2.imread("./landscape Images/color/0.jpg")

# Kim tra xem nh c c c khng
if image is None:
    print("Error: Khng th c c nh.")
else:
    # Chuyn i nh thnh grayscale
    grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # o ngc nh grayscale
    invert = cv2.bitwise_not(grey_image)

    # Lm m nh  o ngc
    blur = cv2.GaussianBlur(invert, (21, 21), 0)

    # o ngc nh  lm m
    inverted_blur = cv2.bitwise_not(blur)

    # To hiu ng tranh v bt ch bng cch chia nh grayscale cho nh lm m o ngc
    sketch = cv2.divide(grey_image, inverted_blur, scale=256.0)

    # Hin th kt qu di dng hnh nh
    plt.imshow(sketch, cmap='gray')  # S dng cmap='gray'  hin th nh di dng grayscale
    plt.axis('off')  # n trc ta 
    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# c nh gc
image = cv2.imread("./landscape Images/color/0.jpg")

# Kim tra xem nh c c c khng
if image is None:
    print("Error: Khng th c c nh.")
else:
    # iu chnh kch thc nh nh li (to hiu ng pixel)
    height, width = image.shape[:2]

    # Gim  phn gii (cng nh, nh s cng pixelated)
    pixel_size = 16  # iu chnh kch thc pixel theo  mun
    temp_image = cv2.resize(image, (width // pixel_size, height // pixel_size), interpolation=cv2.INTER_LINEAR)

    # Phng to li nh v kch thc ban u
    pixelated_image = cv2.resize(temp_image, (width, height), interpolation=cv2.INTER_NEAREST)

    # Chuyn i BGR sang RGB  hin th vi matplotlib
    pixelated_image_rgb = cv2.cvtColor(pixelated_image, cv2.COLOR_BGR2RGB)

    # Hin th nh
    plt.imshow(pixelated_image_rgb)
    plt.axis('off')  # n trc ta 
    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# c nh gc
image = cv2.imread("./landscape Images/color/0.jpg")

if image is None:
    print("Error: Khng th c c nh.")
else:
    # Chuyn thnh grayscale
    grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Lm m nh  gim chi tit
    grey_image = cv2.medianBlur(grey_image, 5)

    # Pht hin cnh bng thut ton adaptive thresholding
    edges = cv2.adaptiveThreshold(grey_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)

    # Lm m v gim  sc nt  to hiu ng hot hnh
    color = cv2.bilateralFilter(image, 9, 300, 300)

    # Kt hp nh mu v cc cnh
    cartoon = cv2.bitwise_and(color, color, mask=edges)

    # Chuyn sang RGB  hin th
    cartoon_rgb = cv2.cvtColor(cartoon, cv2.COLOR_BGR2RGB)

    # Hin th
    plt.imshow(cartoon_rgb)
    plt.axis('off')
    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# c nh gc
image = cv2.imread("./landscape Images/color/0.jpg")

if image is None:
    print("Error: Khng th c c nh.")
else:
    # p dng hiu ng tng cng nh sng CLAHE
    lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)

    # Tng cng knh L bng CLAHE (Contrast Limited Adaptive Histogram Equalization)
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
    l = clahe.apply(l)

    # Gp li thnh nh
    lab = cv2.merge((l, a, b))
    hdr_image = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)

    # Chuyn BGR sang RGB  hin th
    hdr_image_rgb = cv2.cvtColor(hdr_image, cv2.COLOR_BGR2RGB)

    # Hin th
    plt.imshow(hdr_image_rgb)
    plt.axis('off')
    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# c nh gc
image = cv2.imread("./landscape Images/color/0.jpg")

if image is None:
    print("Error: Khng th c c nh.")
else:
    # Chuyn nh sang grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Lm mn nh bng Gaussian Blur
    gray_blur = cv2.medianBlur(gray, 5)
    
    # Pht hin cnh bng thut ton adaptive threshold
    edges = cv2.adaptiveThreshold(
        gray_blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9
    )
    
    # p dng b lc bilateral  lm mn cc vng mu
    color = cv2.bilateralFilter(image, 9, 300, 300)
    
    # Kt hp hnh nh mu v cnh  to hiu ng hot hnh
    cartoon = cv2.bitwise_and(color, color, mask=edges)

    # Chuyn BGR sang RGB  hin th
    cartoon_rgb = cv2.cvtColor(cartoon, cv2.COLOR_BGR2RGB)

    # Hin th
    plt.imshow(cartoon_rgb)
    plt.axis('off')
    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# c nh gc
image = cv2.imread("./landscape Images/color/0.jpg")

if image is None:
    print("Error: Khng th c c nh.")
else:
    # Chuyn i nh sang grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Pht hin cnh bng thut ton Canny
    edges = cv2.Canny(gray, threshold1=100, threshold2=200)

    # Hin th
    plt.imshow(edges, cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# c nh nn v nh cn ghp
src = cv2.imread("./landscape Images/color/0.jpg")  # nh cn ghp
dst = cv2.imread("style_3.png")  # nh nn

# To mask t nh ngun
src_mask = 255 * np.ones(src.shape, src.dtype)

# To im gia nh nn, ni nh ngun s c ghp vo
center = (dst.shape[1] // 2, dst.shape[0] // 2)

# p dng hiu ng ghp Seamless Cloning
output = cv2.seamlessClone(src, dst, src_mask, center, cv2.NORMAL_CLONE)

# Chuyn BGR sang RGB  hin th
output_rgb = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)

# Hin th
plt.imshow(output_rgb)
plt.axis('off')
plt.show()
```
