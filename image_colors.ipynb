```python
import cv2
import matplotlib.pyplot as plt

def apply_negative(image):
    # Kim tra nu nh u vo c hp l khng
    if image is None:
        print("Li: Khng tm thy hoc khng th ti nh")
        return

    # p dng negative
    img_negative = 255 - image

    # Hin nh
    plt.imshow(cv2.cvtColor(img_negative, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Turn off axis
    plt.show()

# Test
image = cv2.imread("./Artistic/data/inp_images/tower.jpeg")
apply_negative(image)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def apply_gray_color(image):
    # Kim tra nu nh u vo c hp l khng
    if image is None:
        print("Li: Khng tm thy hoc khng th ti nh")
        return
    
    # p dng hiu ng mu xm
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Hin th nh kt qu
    plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Tt trc hin th
    plt.show()

# Test
image = cv2.imread("./Artistic/data/inp_images/tower.jpeg")
apply_gray_color(image)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def apply_hue(image, hue_shift):
    # Kim tra nu nh u vo c hp l khng
    if image is None:
        print("Li: Khng tm thy hoc khng th ti nh")
        return

    # Chuyn nh t khng gian mu BGR sang HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Chia knh v p dng gi tr hue_shift
    h = hsv_image[:,:,0]
    s = hsv_image[:,:,1]
    v = hsv_image[:,:,2]
    hsv_image = cv2.add(h, hue_shift)
    hsv_image = cv2.merge([hsv_image,s,v])

    # Chuyn nh t HSV tr li BGR
    result_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)

    # Hin th nh  thay i hue
    plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Tt trc ta 
    plt.show()

# Test
image = cv2.imread("./Artistic/data/inp_images/tower.jpeg")
apply_hue(image, 85)
```
