```python
import cv2
import matplotlib.pyplot as plt

def apply_negative(image):
    # Kiểm tra nếu ảnh đầu vào có hợp lệ không
    if image is None:
        print("Lỗi: Không tìm thấy hoặc không thể tải ảnh")
        return

    # Áp dụng negative
    img_negative = 255 - image

    # Hiện ảnh
    plt.imshow(cv2.cvtColor(img_negative, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Turn off axis
    plt.show()

# Test
image = cv2.imread("./Artistic/data/inp_images/tower.jpeg")
apply_negative(image)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def apply_gray_color(image):
    # Kiểm tra nếu ảnh đầu vào có hợp lệ không
    if image is None:
        print("Lỗi: Không tìm thấy hoặc không thể tải ảnh")
        return
    
    # Áp dụng hiệu ứng màu xám
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Hiển thị ảnh kết quả
    plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Tắt trục hiển thị
    plt.show()

# Test
image = cv2.imread("./Artistic/data/inp_images/tower.jpeg")
apply_gray_color(image)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def apply_hue(image, hue_shift):
    # Kiểm tra nếu ảnh đầu vào có hợp lệ không
    if image is None:
        print("Lỗi: Không tìm thấy hoặc không thể tải ảnh")
        return

    # Chuyển ảnh từ không gian màu BGR sang HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Chia kênh và áp dụng giá trị hue_shift
    h = hsv_image[:,:,0]
    s = hsv_image[:,:,1]
    v = hsv_image[:,:,2]
    hsv_image = cv2.add(h, hue_shift)
    hsv_image = cv2.merge([hsv_image,s,v])

    # Chuyển ảnh từ HSV trở lại BGR
    result_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)

    # Hiển thị ảnh đã thay đổi hue
    plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Tắt trục tọa độ
    plt.show()

# Test
image = cv2.imread("./Artistic/data/inp_images/tower.jpeg")
apply_hue(image, 85)
```
